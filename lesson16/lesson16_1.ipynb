```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

np.random.seed(42)
scores = np.random.randint(50, 101,(50,5))
df1:DataFrame = pd.DataFrame(scores,
             index=[f'學號{i}' for i in range(1,51)],
             columns=["國文","英文","數學","歷史","生物"])
df1
```

```python
#選取1欄
df1['國文']
df1['英文']
```

```python
#選取多欄
#df[list]

df1[['國文','英文']]
```
